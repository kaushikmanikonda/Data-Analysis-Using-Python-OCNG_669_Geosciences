```python
import pandas as pd

url = 'https://waterdata.usgs.gov/tx/nwis/dv?format=rdb&site_no=08108700&referred_module=sw&begin_date=1993-01-01&end_date=2018-10-23'

brazos = pd.read_csv(url, delim_whitespace=True, comment='#', skiprows=[34,], parse_dates=[2, ], index_col=2)
brazos['135330_00060_00003']  # mean daily discharge
```

Explain in the cell below what each of the options does in the function call to `pd.read_csv` in the cell above:

### These descriptions are taken from the "skills" notebook professor posted on canvas.
delim_whitespace=True - Tells the function to use any kind of whitespace (spaces, TABs) to separate the data.

comment='#' - Tells the function that the '#' character indicates a comment and anything after that character, but on the same line, should be ignored.

skiprows=[35] - Indicates that the line 34 should be skipped. Remember that Python starts counting things from 0. So line 36 of the file is at index 35.

parse_dates=[2,] - Indicates that the third column (at index 2) should be run through the automatic date conversion process. The function will do its best to figure out how the date is formatted and convert it. Usually 95% effective. Sometimes the date and time are in different columns. If we put more column numbers in here, it will combine them first, then try to convert them.

index_col=2 - Tells the function to use the third column (at index 2) as the index for the dataframe. This means it will 'sort' the data by date and we can refer to specific dates of data in the future.
